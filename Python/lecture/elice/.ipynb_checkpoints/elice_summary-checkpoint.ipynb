### 월드컵 매치 데이터 전처리
- 저장된 데이터 이상값, 결측값, 변경되어야 하는 값이 있을 경우 정제 필요
- WorldCupMatches.csv 파일에는 일부 이상값이 있기 때문에 전처리 진행
- 데이터 프레임의 일부 값을 대체

```python
world_cups_matches = pd.read_csv("./WorldCupMatches.csv")
world_cups_matches = world_cups_matches.replace("Germany FR", "Germany")
world_cups_matches = world_cups_matches.replace("C�te d'Ivoire", "Côte d'Ivoire")
world_cups_matches = world_cups_matches.replace("rn”>Bosnia and Herzegovina", "Bosnia and Herzegovina")
world_cups_matches = world_cups_matches.replace("rn”>Serbia and Montenegro", "Serbia and Montenegro")
world_cups_matches = world_cups_matches.replace("rn”>Trinidad and Tobago", "Trinidad and Tobago")
world_cups_matches = world_cups_matches.replace("rn”>United Arab Emirates", "United Arab Emirates")
world_cups_matches = world_cups_matches.replace("Soviet Union", "Russia")
```

```python
dupli = world_cups_matches.duplicated()
len(dupli[dupli == True]) 
```

```python
world_cups_matches = world_cups_matches.drop_duplicates()
```

### 국가별 득점 수 구하기
- 역대 월드컵 경기를 통틀어 가장 많은 골을 넣은 국가는 어디였는지 확인
- Year	경기가 진행된 연도	1930
- Datetime	경기 시작 시간	13 Jul 1930 - 15:00
- Stage	스테이지(조별리그, 16강, 8강 등)	Group 1
- Stadium	경기장	Pocitos
- City	도시	Montevideo
- Home Team Name	홈 팀 국가 이름	France
- Home Team Goals	홈 팀 득점 수	4
- Away Team Goals	원정 팀 득점 수	1
- Away Team Name	원정 팀 국가 이름	Mexico
- Win conditions	승리 상황	Italy win after extra time
- Attendance	관중 수	4444
- Half-time Home Goals	홈 팀 전반전 득점 수	3
- Half-time Away Goals	원정 팀 전반전 득점 수	0
- Referee	주심의 이름	LOMBARDI Domingo (URU)
- Assistant 1	제 1 부심(선심)의 이름	CRISTOPHE Henry (BEL)
- Assistant 2	제 2 부심(선심)의 이름	REGO Gilberto (BRA)
- RoundID	라운드의 고유 ID	201
- MatchID	경기의 고유 ID	1096
- Home Team Initials	홈 팀의 세 글자 이니셜	FRA
- Away Team Initials	원정 팀의 세 글자 이니셜	MEX

```python
home = world_cups_matches.groupby('Home Team Name')['Home Team Goals'].sum()
away = world_cups_matches.groupby('Away Team Name')['Away Team Goals'].sum()

goal_per_country = pd.concat([home, away], axis = 1, sort = True).fillna(0)
goal_per_country['goals'] = goal_per_country['Home Team Goals'] + goal_per_country['Away Team Goals']

goal_per_country = goal_per_country['goals'].sort_values(ascending = False)
goal_per_country = goal_per_country.astype(int)
goal_per_country
```

### 역대 월드컵의 국가별 득점 수 - 상위 10개 출력

```python
fig, ax = plt.subplots()
x = goal_per_country.index[:10]
y = goal_per_country[:10]
ax.bar(
  x, y,
  width = 0.5
)
plt.xticks(x, rotation = 30)
plt.tight_layout()
```

### 2014 월드컵 다득점 국가 순위
- 2014 브라질 월드컵에서 각 국가가 기록한 득점 수를 내림차순으로 정렬하여 출력

```python
world_cups_matches_2014 = world_cups_matches[world_cups_matches['Year'] == 2014]
home_team_goals = world_cups_matches_2014.groupby('Home Team Name')['Home Team Goals'].sum()
away_team_goals = world_cups_matches_2014.groupby('Away Team Name')['Away Team Goals'].sum()

team_goals_2014 = pd.concat([home_team_goals, away_team_goals], axis = 1, sort = True).fillna(0)
team_goals_2014['goals'] = team_goals_2014['Home Team Goals'] + team_goals_2014['Away Team Goals']

team_goals_2014 = team_goals_2014.drop(['Home Team Goals', 'Away Team Goals'], axis = 1)
team_goals_2014.astype(int)

team_goals_2014 = team_goals_2014['goals'].sort_values(ascending = False)
print(team_goals_2014)
```

```python
team_goals_2014.plot(x = team_goals_2014.index, y = team_goals_2014.values, kind = 'bar', figsize = (12, 12), fontsize = 10)
```
